```python
import os
from pypdf import PdfReader
from pypdf.errors import PdfReadError, EmptyFileError

folder = "/content/drive/MyDrive/8085_project"

all_text = ""

for file in os.listdir(folder):
    if file.endswith(".pdf"):
        file_path = os.path.join(folder, file)
        print(f"Reading {file}")

        try:
            reader = PdfReader(file_path)
            for page in reader.pages:
                text = page.extract_text()
                if text:
                    all_text += text + "\n"
        except (EmptyFileError, PdfReadError) as e:
            print(f"Error reading {file_path}: {e}. Skipping this file.")
        except Exception as e:
            print(f"An unexpected error occurred while reading {file_path}: {e}. Skipping this file.")

print("Total characters:", len(all_text))
```

```python
def split_text(text, chunk_size=1000):
    chunks = []

    for i in range(0, len(text), chunk_size):
        chunk = text[i:i+chunk_size]

        if len(chunk.strip()) > 100:
            chunks.append(chunk)

    return chunks

chunks = split_text(all_text)

print("Chunks:", len(chunks))
```

### BERT-based Summarization Function

I will now define a function `bert_summarize` that takes a list of text chunks, concatenates them, splits the combined text into sentences using `nltk.sent_tokenize`, generates BERT embeddings for each sentence, and then uses cosine similarity to find sentences most similar to the overall document embedding. These top sentences will form the summary for a given category.

```python
import numpy as np
from nltk.tokenize import sent_tokenize
from sklearn.metrics.pairwise import cosine_similarity

def bert_summarize(chunks_list, top_n=5):
    if not chunks_list:
        return "No relevant information found for this section."

    # Combine all chunks into a single text
    combined_text = " ".join(chunks_list)

    # Split combined text into sentences
    sentences = sent_tokenize(combined_text)
    if not sentences:
        return "No sentences found to summarize."

    # Generate embeddings for each sentence
    sentence_embeddings = []
    for sentence in sentences:
        # BERT can handle max 512 tokens. Truncate if necessary.
        inputs = tokenizer(
            sentence,
            return_tensors="pt",
            truncation=True,
            max_length=512,
            padding=True
        )
        with torch.no_grad():
            outputs = model(**inputs)
        # Use mean of last hidden state as sentence embedding
        embedding = outputs.last_hidden_state.mean(dim=1).squeeze()
        sentence_embeddings.append(embedding.numpy())

    # Convert list of embeddings to a numpy array
    sentence_embeddings_array = np.array(sentence_embeddings)

    # Calculate the document embedding (mean of all sentence embeddings)
    document_embedding = np.mean(sentence_embeddings_array, axis=0)

    # Calculate cosine similarity between document embedding and each sentence embedding
    scores = cosine_similarity(
        [document_embedding],
        sentence_embeddings_array
    )[0]

    # Get indices of top N sentences
    top_indices = np.argsort(scores)[-top_n:]

    # Reconstruct summary in original sentence order
    summary = []
    for idx in sorted(top_indices):
        summary.append(sentences[idx])

    return " ".join(summary)

# Generate summary for Architecture section
architecture_summary = bert_summarize(architecture_chunks)
print("\n--- Architecture Summary ---")
print(architecture_summary)
```
