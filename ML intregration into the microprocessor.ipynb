```python
import pandas as pd

data = {
    "cores": [1,1,1,1,1,2,4,8,16],
    "threads": [1,1,1,1,1,2,4,16,32],
    "frequency": [3,5,8,12,33,66,2000,4000,5000],
    "data_bus": [8,16,16,16,32,32,64,64,64],
    "address_bus": [16,20,20,24,32,32,64,64,64],
    "cache_kb": [0,0,0,1,8,16,512,32768,65536],
    "node_nm": [3000,3000,1500,1500,1000,800,45,7,5],

    "transistors": [
        6500,
        29000,
        29000,
        134000,
        275000,
        1200000,
        230000000,
        10200000000,
        13100000000
    ],

    "power": [
        1.0,
        1.5,
        1.5,
        2.5,
        4.0,
        7.0,
        95,
        105,
        170
    ],

    "performance": [
        5,
        10,
        12,
        18,
        25,
        35,
        60,
        85,
        95
    ]
}

df = pd.DataFrame(data)

df
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
features = [
    "cores",
    "threads",
    "frequency",
    "data_bus",
    "address_bus",
    "cache_kb",
    "node_nm"
]

X = df[features]
```

```python
transistor_model = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

transistor_model.fit(
    X,
    df["transistors"]
)
```

```python
power_model = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

power_model.fit(
    X,
    df["power"]
)
```

```python
performance_model = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

performance_model.fit(
    X,
    df["performance"]
)
```

**Predection Design of Microprocessor**

```python
def predict_design(
    cores,
    threads,
    frequency,
    data_bus,
    address_bus,
    cache_kb,
    node_nm
):

    sample = pd.DataFrame([{
        "cores": cores,
        "threads": threads,
        "frequency": frequency,
        "data_bus": data_bus,
        "address_bus": address_bus,
        "cache_kb": cache_kb,
        "node_nm": node_nm
    }])

    transistors = int(
        transistor_model.predict(sample)[0]
    )

    power = float(
        power_model.predict(sample)[0]
    )

    performance = float(
        performance_model.predict(sample)[0]
    )

    return {
        "transistors": transistors,
        "power": round(power,2),
        "performance": round(performance,2)
    }
```

```python
result = predict_design(
    cores=1,
    threads=1,
    frequency=5,
    data_bus=16,
    address_bus=20,
    cache_kb=0,
    node_nm=3000
)

print(result)
```
